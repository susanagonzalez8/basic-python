### Trabajar con secuencias

La mayoría de las veces cuando pensamos en secuencias tenemos en mente una cadena de caracteres como 'AGTACACTGGT'. Puede crear un objeto Seq con esta secuencia de la siguiente manera:

```python
from Bio.Seq import Seq

my_seq = Seq("AGTACACTGGT")
print(my_seq)
my_seq
```

El objeto Seq difiere de la cadena Python en los métodos que admite. No podríamos hacer esto con una simple cadena:

```python
my_seq
```

```python
my_seq.complement()
```

```python
my_seq.reverse_complement()
```

### Parseo de formatos de archivo de secuencia

Para estos ejemplos será necesario descargar los ficheros [ls_orchid.fasta](https://raw.githubusercontent.com/biopython/biopython/master/Doc/examples/ls_orchid.fasta) y [ls_orchid.gbk](https://raw.githubusercontent.com/biopython/biopython/master/Doc/examples/ls_orchid.gbk)

#### Ejemplo simple de parseo FASTA

Si abre el archivo orchid.fasta en su editor de texto favorito, verá que el archivo comienza así:

>gi|2765658|emb|Z78533.1|CIZ78533 C.irapeanum 5.8S rRNA gene and ITS1 and ITS2 DNA
CGTAACAAGGTTTCCGTAGGTGAACCTGCGGAAGGATCATTGATGAGACCGTGGAATAAACGATCGAGTG
AATCCGGAGGACCGGTGTACTCAGCTCACCGGGGGCATTGCTCCCGTGGTGACCCTGATTTGTTGTTGGG
...

Contiene 94 registros, cada uno tiene una línea que comienza con ">" (símbolo mayor que) seguido de la secuencia en una o más líneas. Ahora probemos esto en Python:

```python
from Bio import SeqIO

for seq_record in SeqIO.parse("ls_orchid.fasta", "fasta"):
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))
```

#### Ejemplo simple de parseo GenBank

Ahora vamos a cargar el fichero GenBank orchid.gbk en su lugar. Tenga en cuenta que el código para hacer esto es casi idéntico al fragmento de código utilizado anteriormente para el archivo FASTA. La única diferencia es que cambiamos el nombre del archivo y la cadena de formato:

```python
from Bio import SeqIO

for seq_record in SeqIO.parse("ls_orchid.gbk", "genbank"):
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))    
```

Notará que se ha utilizado una cadena más corta como seq_record.id en este caso.

### Conexión con bases de datos biológicas

Una de las cosas más comunes que debe hacer en bioinformática es extraer información de bases de datos biológicas. Puede ser bastante tedioso acceder a estas bases de datos manualmente, especialmente si tiene mucho trabajo repetitivo que hacer. Biopython intenta ahorrarle tiempo y energía poniendo a su disposición algunas bases de datos en línea desde scripts de Python. Actualmente, Biopython dispone de código para extraer información de las siguientes bases de datos:

* Entrez (y PubMed)
* ExPASy
* SCOP 

Básicamente, el código de estos módulos facilita la escritura de código de Python que interactúa con estas páginas, de modo que puede obtener resultados en un formato fácil de manejar. En algunos casos, los resultados se pueden integrar estrechamente con los parseadores de Biopython para facilitar aún más la extracción de información.

```python
from Bio import Entrez
#Entrez.email = "A.N.Other@example.com" # Always tell NCBI who you are
handle = Entrez.einfo()
result = handle.read()
handle.close()
```

```python
from Bio import Entrez
Entrez.email = "A.N.Other@example.com" # Always tell NCBI who you are
handle = Entrez.einfo()
result = handle.read()
handle.close()
```
