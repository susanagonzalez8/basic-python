```python
# Promoter motifs are small DNA patterns nearby gene sequences to which the cellular machinery binds in order to help initiate the gene-transcription process.
# the GATA protein binds to any short DNA sequence matching the pattern "[AT]GATA[GA]"

import re

def contar_motifs(secuencia, motif):
    piezas = re.split(motif, secuencia)
    print(piezas)
    return len(piezas) - 1 


secuencia = "AAAAAAAAGATAGAAAAAGATAAAAAATGATAGAAAATGATAAAAAA"
print(contar_motifs(secuencia, "[AT]GATA[GA]"))
```

```python
# Promoter motifs are small DNA patterns nearby gene sequences to which the cellular machinery binds in order to help initiate the gene-transcription process.
# the GATA protein binds to any short DNA sequence matching the pattern "[AT]GATA[GA]"

import re

def count_motifs(seq, motif):
    pieces = re.split(motif, seq)
    return len(pieces) - 1 


file = open("grape_promoters.txt", "r")

for linea in file :
    linea = linea.strip()
    lista = re.split("\s+", linea)
    gid = lista[0]
    seq = lista[1]
    
    num_motifs = count_motifs(seq, "[AT]GATA[GA]")
    print(gid + "\t" + str(num_motifs))

file.close()
```
