## <font color='FIREBRICK'>Ejercicio 1 (7 puntos)</font>

<font size='3px'><p style="text-align: justify;line-height: 150%">Queremos almacenar los datos de nuestros actores favoritos en una base de datos (BBDD), pero no nos gusta ninguna de las disponibles, por lo tanto vamos a construirla nosotros mismos.</p></font>
    
<font size='3px'><p style="text-align: justify;line-height: 150%">La BBDD será un diccionario de diccionarios. Para el primer diccionario la clave será el nombre del actor. El segundo diccionario (el que estaría dentro del primer diccionario) consta de 3 campos: año de nacimiento, sexo y una lista con el nombre de las películas en las que ha participado.</p></font>

<font size='3px'><p style="text-align: justify;line-height: 150%">El programa nos permitirá hacer una serie de cosas, elegidas por un menú:</p></font>

<div style="font-size: 16px; line-height: 150%">
<ol>
    1. Introducir los datos de un nuevo actor o actriz (el número de películas puede variar de un actor a otro) (si el actor ya existía en la BBDD se sobreescribirán los datos).<br/>
    2. Listar todos los actores mostrando únicamente sus nombres (con cierto formato).<br/>
    3. Mostrar los datos de un determinado actor o actriz (con cierto formato).<br/>
    4. Buscar aquellos actores cuyo año de nacimiento se encuentre en un determinado rango de años (que se le pedirá al usuario).<br/>
    5. Buscar aquellos actores de un sexo determinado.<br/>
    6. Buscar aquellos actores que hayan participado en una película.<br/>
    0. Salir del programa.        
</ol>
</div>

```python
# Realice el ejercicio dentro del cuadro a continuación de esta línea 

# Os dejo un ejemplo de inicialización de la base de datos para que veáis la estructura (vuestra base de datos inicialmente estará vacía).

BBDD_actores = {
    'Keanu Reeves' : {
        'año' : 1964, 
        'sexo' : 'masculino', 
        'peliculas' : ['Speed', 'The Matrix', 'Constantine']
    },
    'cameron diaz' : {
        'año' : 1982, 
        'sexo' : 'femenino', 
        'peliculas' : [ 'The Matrix', 'Constantine']
    },
    'brad pitt' : {
        'año' : 2012, 
        'sexo' : 'masculino', 
        'peliculas' : ['Speed', 'The Matrix', 'Constantine']
    }
}   

print ('''
1. Introducir los datos de un nuevo actor o actriz (el número de películas puede variar de un actor a otro) (si el actor ya existía en la BBDD se sobreescribirán los datos).
2. Listar todos los actores mostrando únicamente sus nombres (con cierto formato).
3. Mostrar los datos de un determinado actor o actriz (con cierto formato).
4. Buscar aquellos actores cuyo año de nacimiento se encuentre en un determinado rango de años (que se le pedirá al usuario).
5. Buscar aquellos actores de un sexo determinado.
6. Buscar aquellos actores que hayan participado en una película.
0. Salir del programa.
''')


while True: 
    seleccion = input ("¿Qué quieres hacer? ")

    if seleccion == "1":
        actor = input("Introduzca un actor: ")
        year = int (input ("Introduzca el año: "))
        sexo = input("Introduzca el sexo: ")
        peliculas = input("Introduzca las peliculas separadas por comas: ").split(",")
        dic = {
            actor : {
                'año' : year,
                'sexo': sexo, 
                'peliculas': peliculas
            }
        }
        BBDD_actores.update(dic)
        print(BBDD_actores)
        
    elif seleccion == "2":
        for actor in BBDD_actores:
            txt = "El nombre del actor/actriz es {}".format(actor)
            print(txt)
        
    elif seleccion == "3":
        actor = input ("Introduzca un actor: ")
        year = BBDD_actores[actor]['año']
        sexo = BBDD_actores[actor]['sexo']
        peliculas = BBDD_actores[actor]['peliculas']
        txt = "El nombre del actor es {actor}. \n\tnacio en el año {year}, \n\tel sexo es {sexo} \n\tsus peliculas son: {peliculas}".format(actor = actor, year = year, sexo = sexo, peliculas = ", ".join(peliculas))
        print(txt)
        
    elif seleccion == "4":
        year_1 = int (input ("Introduzca la horquilla inferior: "))
        year_2 = int (input ("Introduzca la horquilla superior: "))
        for actor in BBDD_actores:
            if BBDD_actores[actor]['año'] > year_1 and BBDD_actores[actor]['año'] < year_2:
                print(actor)
            
    elif seleccion == "5":
        sexo = input ("Introduzca un sexo: ")
        for actor in BBDD_actores:
            if BBDD_actores[actor]['sexo'] == sexo:
                print(actor)
        
    elif seleccion == "6":
        pelicula = input ("Introduzca una pelicula: ")
        for actor in BBDD_actores:
            if pelicula in BBDD_actores[actor]['peliculas']:
                print(actor)
    
    elif seleccion == "0":
        break
        
    else:
        print('opcion incorrecta')
    
```
